# 追踪数据转换测试 - 2022世界杯决赛

本notebook用于测试追踪数据转换功能在2022世界杯决赛数据上的运行

**比赛**: 阿根廷 vs 法国 (Game ID: 10517)

**基于**: 已验证可运行的 `test_convert_tracking_final.py`

**测试内容**:
1. 获取可用比赛列表
2. 加载决赛元数据
3. 处理追踪数据
4. 数据统计和验证

## 1. 导入必要的库

```python
import sys
import os

# 添加父目录到Python路径（用于导入项目模块）
parent_dir = os.path.abspath(os.path.join(os.getcwd(), '..'))
if parent_dir not in sys.path:
    sys.path.insert(0, parent_dir)
    print(f"✓ 已添加父目录到Python路径: {parent_dir}")

# 导入自定义模块
from convert_tracking import process_game, get_available_games, load_game_metadata

print("✓ 所有库导入成功")
```

## 2. 配置参数

```python
# 2022世界杯决赛ID
FINAL_GAME_ID = '10517'

print("="*80)
print("2022 FIFA World Cup Final - Data Processing Test")
print("="*80)
print(f"\n使用比赛ID: {FINAL_GAME_ID}")
print(f"输出目录: Data/{FINAL_GAME_ID}/")
```

## 3. 步骤1: 获取所有可用比赛

```python
print("\n步骤1: 获取所有可用比赛...")

try:
    games = get_available_games()
    print(f"[OK] 找到 {len(games)} 场比赛")
    print(f"\n比赛ID列表:")
    for i, game_id in enumerate(games, 1):
        print(f"  {i:2d}. {game_id}")
        if i >= 10:  # 只显示前10场
            print(f"  ... 还有 {len(games) - 10} 场比赛")
            break
    
    # 检查决赛ID是否存在
    if FINAL_GAME_ID in games:
        print(f"\n✓ 决赛比赛ID {FINAL_GAME_ID} 已找到")
    else:
        print(f"\n✗ 错误: 决赛比赛ID {FINAL_GAME_ID} 未找到")
        print(f"  可用比赛: {games}")
        
except Exception as e:
    print(f"[ERROR] {e}")
    import traceback
    traceback.print_exc()
```

## 4. 步骤2: 加载决赛元数据

```python
print(f"\n步骤2: 加载决赛元数据 (ID: {FINAL_GAME_ID})...")

try:
    metadata = load_game_metadata(FINAL_GAME_ID)
    print(f"[OK] 元数据加载成功\n")
    
    print("="*80)
    print("比赛信息")
    print("="*80)
    print(f"主队: {metadata['homeTeam']['name']}")
    print(f"客队: {metadata['awayTeam']['name']}")
    print(f"日期: {metadata['date']}")
    print(f"球场: {metadata['stadium']['name']}")
    print(f"主队开球方向: {'左' if metadata['homeTeamStartLeft'] else '右'}")
    
except Exception as e:
    print(f"[ERROR] {e}")
    import traceback
    traceback.print_exc()
```

## 5. 步骤3: 处理决赛追踪数据

**注意**: 这个过程可能需要2-5分钟时间。

```python
print(f"\n步骤3: 处理决赛追踪数据...")
print("注意: 这可能需要2-5分钟时间...\n")

try:
    balls_df, events_df, players_df = process_game(FINAL_GAME_ID, save_output=True)
    
    print(f"\n[OK] 决赛数据处理成功！")
    print("\n" + "="*80)
    print("数据统计")
    print("="*80)
    
except Exception as e:
    print(f"\n[ERROR] 处理失败: {e}")
    import traceback
    traceback.print_exc()
```

## 6. 球数据统计

```python
print("\n[球数据]")
print(f"  总帧数: {len(balls_df):,}")
print(f"  时间范围: {balls_df['periodElapsedTime'].min():.2f}s - {balls_df['periodElapsedTime'].max():.2f}s")
print(f"  比赛阶段: {sorted(balls_df['period'].unique())}")

print(f"\n  前5行数据:")
print(balls_df[['frameNum', 'period', 'periodElapsedTime', 'x', 'y', 'z']].head())
```

## 7. 事件数据统计

```python
print("\n[事件数据]")
print(f"  总事件数: {len(events_df):,}")

if len(events_df) > 0:
    print(f"  事件类型数: {events_df['eventType'].nunique()}")
    print(f"  涉及球队: {list(events_df['team_name'].unique())}")
    
    print(f"\n  事件类型分布:")
    event_counts = events_df['eventType'].value_counts()
    for event_type, count in event_counts.head(10).items():
        print(f"    {event_type}: {count}")
    
    print(f"\n  前5个事件:")
    print(events_df[['periodElapsedTime', 'eventType', 'player_name', 'team_name']].head())
```

## 8. 球员数据统计

```python
print("\n[球员数据]")
print(f"  总记录数: {len(players_df):,}")
print(f"  球员数量: {players_df['playerName'].nunique()}")
print(f"  球队: {list(players_df['teamName'].unique())}")

# 按球队统计
for team in players_df['teamName'].unique():
    team_players = players_df[players_df['teamName'] == team]['playerName'].unique()
    print(f"\n  {team} ({len(team_players)} 名球员):")
    
    # 显示首发11人
    for i, player in enumerate(sorted(team_players)[:11], 1):
        player_data = players_df[players_df['playerName'] == player].iloc[0]
        pos = player_data['playerPos']
        jersey = int(player_data['jerseyNum'])
        print(f"    {i:2d}. #{jersey:2d} {player:20s} ({pos})")
```

## 9. 输出文件验证

```python
print("\n" + "="*80)
print("输出文件")
print("="*80)

output_dir = f"Data/{FINAL_GAME_ID}"
print(f"  目录: {output_dir}/")

files = [
    f"balls_{FINAL_GAME_ID}.csv",
    f"events_{FINAL_GAME_ID}.csv",
    f"players_{FINAL_GAME_ID}.csv"
]

for filename in files:
    filepath = os.path.join(output_dir, filename)
    if os.path.exists(filepath):
        file_size = os.path.getsize(filepath) / (1024 * 1024)  # MB
        print(f"  ✓ {filename} ({file_size:.2f} MB)")
    else:
        print(f"  ✗ {filename} (不存在)")
```

## 10. 数据质量检查

```python
print("\n" + "="*80)
print("数据质量检查")
print("="*80)

# 检查每帧球员数量
print("\n1. 每帧球员数量")
frame_player_counts = players_df.groupby('frameNum').size()
print(f"  最少: {frame_player_counts.min()} 人/帧")
print(f"  最多: {frame_player_counts.max()} 人/帧")
print(f"  平均: {frame_player_counts.mean():.1f} 人/帧")
print(f"  标准(22人)帧数: {(frame_player_counts == 22).sum():,}")

# 检查坐标范围
print("\n2. 坐标范围 (标准球场: 0-105m × 0-68m)")
print(f"  球员X: [{players_df['x'].min():.2f}, {players_df['x'].max():.2f}]")
print(f"  球员Y: [{players_df['y'].min():.2f}, {players_df['y'].max():.2f}]")
print(f"  球X: [{balls_df['x'].min():.2f}, {balls_df['x'].max():.2f}]")
print(f"  球Y: [{balls_df['y'].min():.2f}, {balls_df['y'].max():.2f}]")

# 检查速度
if 'velocity_magnitude' in players_df.columns:
    print("\n3. 速度统计 (m/s)")
    print(f"  球员平均速度: {players_df['velocity_magnitude'].mean():.2f}")
    print(f"  球员最高速度: {players_df['velocity_magnitude'].max():.2f}")
    if 'velocity_magnitude' in balls_df.columns:
        print(f"  球平均速度: {balls_df['velocity_magnitude'].mean():.2f}")
        print(f"  球最高速度: {balls_df['velocity_magnitude'].max():.2f}")

# 检查缺失值
print("\n4. 数据完整性")
print(f"  球员数据缺失值: {players_df.isnull().sum().sum()}")
print(f"  球数据缺失值: {balls_df.isnull().sum().sum()}")
print(f"  事件数据缺失值: {events_df.isnull().sum().sum()}")
```
